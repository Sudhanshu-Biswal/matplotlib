# Pie Chart

```python
#simple data
classes = ["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_student = [45,15,35,25,30]
```

```python
plt.pie(class1_student,labels=classes,autopct="%0.1f%%")
plt.show()
```

```python
explode = [0.03,0,0.1,0,0]
colors = ["c","b","r","y","g"]
textprops={"fontsize":15}
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,
        startangle=270,textprops=textprops,explode=explode)
plt.show()
```

```python
plt.figure(figsize=(16,9))
wedgeprops = {"linewidth":4,"width":3,"edgecolor":"k"}
plt.pie(
    class1_student,
    explode=explode,
    labels=classes,
    colors=colors,
    autopct="%0.2f%%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.4,
    startangle=270,
    radius=0.5,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=True
)
plt.legend()
plt.show()
```

# 3D surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z = np.log(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

# Contour Plots

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```
